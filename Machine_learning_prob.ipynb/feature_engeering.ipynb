# Feature Engeering

```python
import pandas as pd 
room_length = [18,20,10,12,18,11]
room_breadth = [20,20,10,11,19,10]
room_type = ['Big','Big','Normal','Normal','Big','Normal']
data = pd.DataFrame({'Length':room_length,'Breadth':room_breadth,'type':room_type})
data
```

```python
data['Area'] = data['Length']*data['Breadth']
data
```

# Dummy coding categorical (nominal) variables

```python
age = [18,20,23,19,18,22]
city = ['City A','City B','City B','City A','City C','City B']
data = pd.DataFrame({'age':age,'city':city})
data
```

```python
dummy_feature = pd.get_dummies(data['city'])
```

```python
dummy_feature
```

```python
data_age = pd.DataFrame(data=data,columns=['age'])
data_mod = pd.concat([data_age.reset_index(drop=True),dummy_feature],axis=1)
data_mod
```

 # Encoding the categorical (ordinal) variables
 ##### label encoder of sklearn.preprocessing library can be used to encode categorical variable 

```python
import pandas as pd
from sklearn import preprocessing
marks_math = [75,62,90,95,42,57]
marks_science = [78,56,87,91,45,62]
grade = ['B','C','A','A','D','B']
data = pd.DataFrame({'Science marks':marks_science,'Maths marks':marks_math,'Total_grade':grade})
data
```

```python
# option 1
le = preprocessing.LabelEncoder()
le.fit(data['Total_grade'])
data['Total_grade'] = le.transform(data['Total_grade'])
data
```

```python
# option 2 
target = data['Total_grade'].replace(['A','B','C','D'],[0,1,2,3])
target
predictor =data.iloc[:,0:2]
data_mod = pd.concat([predictor.reset_index(drop=True),target],axis=1)
data_mod
```

# Transforming the Numeric data (continuous) feature to categorical feature

```python
import pandas as pd
apartment_area = [4720,2430,4368,3969,6142,7912]
apartment_price =[2360000,1215000,2184000,1984500,3071000,3956000]
data = pd.DataFrame({'Area':apartment_area,'Price':apartment_price})
data
```

```python
import numpy as np
data['Price'] = np.where(data['Price']>3000000,'High',np.where(data['Price']<2000000,'Low','Medium'))
data
```
